1. Read the [population](https://datascience.quantecon.org/assets/data/wdi_population.csv) dataset and [financial](https://datascience.quantecon.org/assets/data/wdi_data.csv) dataset into separate dataframes. No need to change the index names.

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_pop = pd.read_csv('https://datascience.quantecon.org/assets/data/wdi_population.csv')
df_fin = pd.read_csv('https://datascience.quantecon.org/assets/data/wdi_data.csv')

df_pop.head()
```

2. Create a slice of financial dataset for the year 2017 rows, and create another slice for the year 2000 rows; store them in separate variables called 'fin2017' and 'fin2000' respectively.

 Use the above two variables to creae ONE line plot with two lines; one displays the GDP of the countries for 2017 and the other one displays the GDP of the countries for 2000. Make sure to:

*   set the x axis as country names and y axis as GDP
*   set proper linestyle and marker for each line
*   set proper color for line and markers
*   set proper legend for the plot with labels for each line
*   set proper plot title and axes labels


```
df_fin2017 = df_fin[df_fin['year'] == 2017]
df_fin2000 = df_fin[df_fin['year'] == 2000]

# Plot data for 2017
plt.plot(df_fin2017['country'], df_fin2017['GDP'], linestyle='-', marker='o', color='blue', label='2017')

# Plot data for 2000
plt.plot(df_fin2000['country'], df_fin2000['GDP'], linestyle='--', marker='s', color='red', label='2000')

# Add labels, title, legend, etc.
plt.legend()
plt.title("Country's GDPs in 2017 vs 2000")
plt.xlabel('Country Names')
plt.ylabel('GDP')

# Improve layout and show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. Slice the population dataset for the year 2017 records, and store the slice in 'pop2017' variable. Next, create a bar plot to compare the population of the countries in 2017. Make sure to set proper bar color, plot title, and axes labels.

```
pop2017 = df_pop[df_pop['year'] == 2017]
pop2017.head()

plt.bar(pop2017['country'], pop2017['Population'], color='orange')
plt.title('Population of Different Countries in 2017')
plt.xlabel('Country Names')
plt.ylabel('Population (in millions)')
```

4. Slice the financial dataset for the rows for the United States. Then, use that to create a scatter plot to display the 'Exports' against the 'Imports' of the United States, and use the 'year' as color for the markers; it means the markers (data points) of later years will be shown with brighter colors on the scatterplot.

```
US_fin = df_fin[df_fin['country'] == 'United States']
US_fin.head()

plt.scatter(US_fin['Exports'], US_fin['Imports'], c=US_fin['year'], cmap='viridis')

# Add x, y labels, title, and colorbar
plt.xlabel('Exports')
plt.ylabel('Imports')
plt.title('Exports vs Imports in the United States from 2000 to 2016')
plt.colorbar()

plt.show()
```

5. Use pop2017 dataframe to create a pie chart that displays the countries' population as a portion of the total population. Make sure to properly label the portions of the pie chart with country names, and display the percentage of population they hold (the percentage should be float with one decimal place and a % sign). Explode the 'Canada' portion of the pie chart and properly title the pie plot.

```
explode = [0.1 if country == 'Canada' else 0 for country in pop2017['country']]
colors = plt.cm.Paired(range(len(pop2017)))

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    pop2017['Population'],
    labels=pop2017['country'],
    autopct=lambda pct: f'{pct:.1f}%',  # Format percentage
    explode=explode,
    colors=colors,
    startangle=140,
    wedgeprops={'edgecolor': 'black'}  # Add border for better visibility
)

# Title
plt.title('Population Distribution by Country (2017)')

# Show the chart
plt.show()
```
