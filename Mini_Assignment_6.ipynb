2. Check if there are any duplicate records (rows). How many duplicate records are there?

```
duplicate_records = df.duplicated()
duplicate_records.sum()
#there are no duplicate records
```

3. Display the ratio (in percentage) of missing values for each column. Then drop any column that has over 50% of missing values. Next, do the same for the rows; meaning drop the rows with over 50% missing values. Display the number of rows before and after you drop such rows.

```
#show missing percentage for columns
missing_col_percentage = df.isnull().mean()*100
missing_col_percentage
```

```
#drop columns with more than 50% NA values
df = df.loc[:, missing_col_percentage <= 0.5]
df.columns
#removed 'Pacific' and 'Transit' columns that were >50%
```

```
#missing rows
missing_row_percentage = df.isnull().mean(axis=1)*100
missing_row_percentage
```

```
df.shape[0]
```

```
#drop missing rows
df = df.loc[missing_row_percentage <=0.5]
df.shape[0]
#dropped 38 rows
```

4. For the remaining missing values in the dataframe, fill them all with the median value of its column.

```
df = df.fillna(df.median(numeric_only=True))
df
```
